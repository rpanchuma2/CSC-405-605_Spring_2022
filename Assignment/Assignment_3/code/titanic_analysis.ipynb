We are going to study the survival rate of passengers on titanic and what variables affected survival.

Load the dataset in `titanic.xls`. It contains data on all the passengers that travelled on the Titanic.

```python
import pandas as pd

# you would need xlrd - pip install xlrd
t_file = pd.ExcelFile('../data/titanic.xls')
t_df = t_file.parse("titanic", header=0)
t_df.head()
```

### Women and children first?

*** 1. Use the `groupby` method to calculate the proportion of passengers that survived by sex. (1 pt)*** 

```python
#using groupby and using condition survived == 1
survived_people = t_df[t_df["survived"] == 1]
survived_people = survived_people.groupby(["sex","survived"])
survived_people.first()
```

*** 2. Calculate the same proportion, but by class and sex. (2 pts)*** 

```python
#using groupby with multiple variables like pclass , sex
survived_people = t_df[t_df["survived"] == 1]
survived_people = survived_people.groupby(["sex","survived","pclass"])
survived_people.first()
```

*** 3. Create age categories: children (under 14 years), adolescents (14-20), adult (21-64), and senior(65+), and calculate survival proportions by age category, class and sex. (3.5 pts)***

#created a dataframe for each of the age category using the condition age and survived == 1

```python
titanic_survived = t_df[t_df["survived"] == 1]
titanic_survived = titanic_survived[titanic_survived["age"]<=14]
children = titanic_survived.groupby(["age","pclass","sex"])
children.first()



```

adolescents dataframe for those who have survived and groupby sex, class, age category

```python
titanic_survived = t_df[t_df["survived"] == 1]
titanic_survived = titanic_survived[(titanic_survived["age"]>=14) & (titanic_survived["age"]<=20)]
adolescents = titanic_survived.groupby(["age","pclass","sex"])

adolescents.first()
```

adult dataframe for those who have survived and groupby sex, class, age category

```python
titanic_survived = t_df[t_df["survived"] == 1]
titanic_survived = titanic_survived[(titanic_survived["age"]>=20) & (titanic_survived["age"]<=65)]
adult = titanic_survived.groupby(["age","pclass","sex"])
adult.first()
```

senior dataframe for those who have survived and groupby sex, class, age category

```python
titanic_survived = t_df[t_df["survived"] == 1]
titanic_survived = titanic_survived[(titanic_survived["age"]>65)]
senior = titanic_survived.groupby(["age","pclass","sex"])
senior.first()
```
